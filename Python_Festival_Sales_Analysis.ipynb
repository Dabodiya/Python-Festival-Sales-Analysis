```python
ax = snb.countplot(x = 'Gender', data = df, hue = 'Gender', palette = 'Set2')
```

<span style="font-size:24px; font-weight:bold;">Exploratory Data Analysis</span>


<span style="font-size:20px; font-weight:bold;">Gender</span>


```python
ax = snb.countplot(x = 'Gender', data = df, hue = 'Gender', palette = 'Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

snb.barplot(x = 'Gender' , y = 'Amount', data = sales_gen, hue = 'Gender', palette = 'Set1')
```

<span style = "font-size:18px">From Above Graphs we can see that most of the buyers are females and even the purchasing power of females are greater than mens. </span>

<span style="font-size:20px; font-weight:bold;">Age</span>


```python
ax = snb.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

snb.barplot(x = 'Age Group' , y = 'Amount', data = sales_age, hue = 'Age Group', palette = 'Set2')
```

<span style="font-size:18px;">From Above Graph we can see that most of the buyers are of age group between 26 - 35 yrs female.</span>


<span style="font-size:24px; font-weight:bold;">State</span>


```python
#Total no of orders from top 10 states 
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

snb.set(rc = {'figure.figsize' : (17,5)})
snb.barplot(data = sales_state, x = 'State', y = 'Orders', hue = 'State', palette = 'Set2')
```

```python
#Total sales from top 10 states
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

snb.set(rc = {'figure.figsize' : (17,5)})
snb.barplot(data = sales_state, x = 'State', y = 'Amount', hue = 'State', palette = 'Set2')
```

<span style="font-size:18px; ">From above graph we can see that most of the Orders and Sales are from Uttar Pradesh, Maharastra and Karnatka respectively.</span>


<span style="font-size:20px; font-weight:bold;">Marital Status</span>


```python
ax = snb.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status', palette = 'Set2')

snb.set(rc = {'figure.figsize' : (7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

snb.set(rc = {'figure.figsize' : (6,5)})
snb.barplot(data = sales_state, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

<span style="font-size:18px;">From Above Graphs we can see that most of the buyers are married (women) and they have high puchasing power.</span>


<span style="font-size:20px; font-weight:bold;">Occupation</span>


```python
snb.set(rc = {'figure.figsize' : (20,5)})
ax = snb.countplot(data = df, x = 'Occupation', hue = 'Occupation', palette = 'Set1')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occu = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

snb.set(rc = {'figure.figsize' : (20,5)})
snb.barplot(data = sales_occu, x = 'Occupation', y = 'Amount', hue = 'Occupation', palette = 'Set2')
```

<span style="font-size:18px;">From above graph we can see that most of the buyers are working in IT, Aviation and Heathcare sector.</span>


<span style="font-size:20px; font-weight:bold;">Product Category</span>


```python
snb.set(rc = {'figure.figsize' : (20,5)})
ax = snb.countplot(data = df, x = 'Product_Category', hue='Product_Category', palette = 'Set2')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_procat = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)

snb.set(rc = {'figure.figsize' : (20,5)})
snb.barplot(data = sales_procat, x = 'Product_Category', y = 'Amount', hue= 'Product_Category', palette = 'Set2')
```

<span style="font-size:18px;">From above graph we can see that most of the products are sold from Food, Clothing, Electronics and Footwears category.</span>

```python
sales_procat = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)

snb.set(rc = {'figure.figsize' : (20,5)})
snb.barplot(data = sales_procat, x = 'Product_ID', y = 'Orders', hue= 'Product_ID', palette = 'Set1')
```
